```python
# Coverting a dictionary into dataframe--- (no. of values insisde dict value pair should be same otherwise error arises.)

d1 = {"key1" : [1,2,3,4,5,"anand"],  #<----- value inside 'key' must be of same number
      "key2" : [7,8,9,10,11,12],
      "key3" : [13,14,15,16,17,18],
      "key4" : [19,20,21,22,23,24]
}
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(d1)
```

```python
df1
```

```python
pd.DataFrame(d1, index=['a','b','c','d','e','f'])
```

```python
d2 = {"key1" : [1,2,3,4,5,"anand"],  
      "key2" : [7,8,"kumar",10,11,12],
      "key3" : [13,14,15,16,17,18],
      "key4" : [19,20,21,"ineuron","pwskills",24]
}
```

```python
df2=pd.DataFrame(d2)
```

```python
df2
```

```python
# Merge opertion:---
```

```python
pd.merge(df1,df2) #inner join values will come means the coommon values in both dataframe.
```

```python
pd.merge(df1,df2, how= 'inner', on='key1')
```

```python
pd.merge(df1,df2, how= 'inner', on='key4')
```

```python
pd.merge(df1,df2, how= 'left', on='key4') #it will take all data from left side
```

```python
pd.merge(df1,df2, how= 'right', on='key4')
```

```python
# outer combination:- it will bring everyhting fom left and right both
```

```python
pd.merge(df1,df2, how= 'outer', on='key4')
```

```python
pd.merge(df1,df2, how= 'outer', on=['key4','key1'])
```
