```python
# What are the top 5 most installed apps in each popular category?? -- Multivariate  analysis
```

```python
df.head(2)
```

```python
dfa = df.groupby(['Category','App'])['Installs'].sum().reset_index()
```

```python
dfb = dfa.sort_values('Installs', ascending=False)
```

```python
dfb
```

```python
dfb[dfb.Category=='GAME'][:5]
```

```python
dfb[dfb.Category=='FAMILY'][:5]
```

```python
dfb[dfb.Category=='PRODUCTIVITY'][:5]
```

```python
dfb[dfb.Category=='SOCIAL'][:5]
```

```python
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfb[dfb.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
